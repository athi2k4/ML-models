```python
import time
import keras_cv
from tensorflow import keras
import matplotlib.pyplot as plt

# Ensure you're using the right KerasCV version that includes StableDiffusion
model = keras_cv.models.StableDiffusion(img_width=512, img_height=512)

# Generate images
images = model.text_to_image("astronaut riding a horse", batch_size=1)

# The images generated by text_to_image might be in float32 format, so convert to uint8
images = (images * 255).astype("uint8")

def plot_images(images):
    plt.figure(figsize=(20, 20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")

# Plot the generated images
plot_images(images)
plt.show()
```
